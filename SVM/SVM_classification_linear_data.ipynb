```python
import pandas as pd
# imoprt numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.drop('SNo' , axis=1 , inplace=True)
```

```python
df.columns = ["A" , "B" , "C"]
```

```python
# df.drop('SNo',axis=1,inplace=True)
```

```python
sns.scatterplot(x='A' , y='B' , data=df , hue='C')
plt.show()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# prompt: generate the SVM model , train and test the data set and return the accuracy report

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model = SVC()
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
print(accuracy_score(y_test , y_pred))
```

```python
# prompt: mltxtend grapgh for the above model and x y data set

import matplotlib.pyplot as plt
import numpy as np
# !pip install mlxtend

from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X=x.values,
                      y=y.values.astype(np.integer),
                      clf=model,
                      legend=2)

# Update plot object with X/Y axis labels and Figure Title
plt.xlabel(x.columns[0], size=14)
plt.ylabel(x.columns[1], size=14)
plt.title('SVM Decision Region Boundary', size=16)
```

# model with linear kernal

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
model_li = SVC(kernel='linear')
model_li.fit(x_train,y_train)
y_pred = model_li.predict(x_test)
print((accuracy_score(y_test , y_pred))*100)
```

```python
model_li.score(x_train,y_train)*100
```

```python
model_li.score(x_test,y_test)*100
```

this model is underfitting

```python
import matplotlib.pyplot as plt
import numpy as np
# !pip install mlxtend

from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X=x.values,
                      y=y.values.astype(np.integer),
                      clf=model_li,
                      legend=2)

# Update plot object with X/Y axis labels and Figure Title
plt.xlabel(x.columns[0], size=14)
plt.ylabel(x.columns[1], size=14)
plt.title('SVM Decision Region Boundary', size=16)
```
