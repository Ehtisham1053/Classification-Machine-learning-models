```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Purchase_new.csv')
```

```python
df.head()
```

```python
# prompt: convert all the yes to 1 and no to 0 in df

df = df.replace({'Yes':1, 'No':0 , 'yes':1 , 'no':0})
df.head()
```

```python
df.isnull().sum()
```

```python
# prompt: draw the scatter plot for the above dataset and hue parameter as purchase

import matplotlib.pyplot as plt
sns.pairplot(df, hue='Purchase')
plt.show()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# prompt: generate the decision tree classfication using entropy and train and test the model using the dataset and return the accuracy report

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create the decision tree classifier
dtree_entropy = DecisionTreeClassifier(criterion='entropy')

# Train the model
dtree_entropy.fit(x_train, y_train)

# Make predictions on the test set
y_pred_entropy = dtree_entropy.predict(x_test)

# Calculate accuracy
accuracy_entropy = accuracy_score(y_test, y_pred_entropy)
print("Accuracy (Entropy):", accuracy_entropy)
```

```python
# prompt: plot the plot_tree for the  above model

import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtree_entropy, filled=True)
plt.show()
```

now generate the model using the gini index


```python
# prompt: generate the decision tree model using the ginni index , test and train the model and return the accuracy report

# Create the decision tree classifier using Gini index
dtree_gini = DecisionTreeClassifier(criterion='gini')

# Train the model
dtree_gini.fit(x_train, y_train)

# Make predictions on the test set
y_pred_gini = dtree_gini.predict(x_test)

# Calculate accuracy
accuracy_gini = accuracy_score(y_test, y_pred_gini)
print("Accuracy (Gini):", accuracy_gini)
```

```python
# prompt: plot_tree for ginni model

import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(dtree_gini, filled=True)
plt.show()
```
