```python
df = pd.read_csv('KNNAlgorithmDataset.csv')
```

```python
df.drop(columns=['id' , "Unnamed: 32"], inplace=True)
```

```python
df.replace({'M': 1, 'B': 0}, inplace=True)
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0]
```

```python
x.head()
```

```python
y.head()
```

```python
x.iloc[0,:]
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=42)
```

```python
# prompt: generate the KNN model and train , test the model and return the accuracy report and tell is this overfitting or underfitting

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)

y_pred_train = knn.predict(x_train)
y_pred_test = knn.predict(x_test)

accuracy_train = accuracy_score(y_train, y_pred_train)
accuracy_test = accuracy_score(y_test, y_pred_test)

print("Training Accuracy:", accuracy_train)
print("Testing Accuracy:", accuracy_test)

if accuracy_train > accuracy_test:
  if accuracy_train - accuracy_test > 0.1:
    print("Model is overfitting.")
  else:
    print("Model is slightly overfitting.")
elif accuracy_train < accuracy_test:
  print("Model is underfitting.")
else:
  print("Model is perfectly fit.")
```
